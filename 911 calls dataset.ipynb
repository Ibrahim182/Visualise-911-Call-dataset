## Dataset source: https://www.kaggle.com/mchirico/montcoalert

```python
import pandas as pd
import seaborn as sns

sns.set(rc={'figure.figsize': [15, 15]}, font_scale=1.2)
```

```python
# load the data set
df = pd.read_csv('datasets/911.csv')
# take a look at the data
df.head()
```

## which reason is reported more

```python
sns.catplot(x='Reason', data=df, kind='count', palette='Set3');
```

## which hour has the most calls

```python
sns.catplot(x='hour', data=df, kind='count', palette='pastel', height=8);
```

## which month has the most calls

```python
sns.catplot(x='month', data=df, kind='count', palette='Paired', height=6);
```

## which days of the week has the most calls

```python
sns.catplot(x='day of week', data=df, kind='count', palette='Set3', height=6);
```

## which year has the most calls

```python
sns.catplot(x='Year', data=df, kind='count', palette='Paired');
```

## for each hour which reson was reported more

```python
sns.catplot(x='hour',hue='Reason', data=df, kind='count', palette='Paired', height=10);
```
